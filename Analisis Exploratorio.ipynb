# Evolución precio m2 en dólares por mes a través de los años

```python
#Calculamos el valor promedio del m2 en dolares por mes para el año 2013
#Primero nos aseguramos que en el frame haya datos de solo el 2013

props_2013_usd_month = props_2013[(props_2013['created_on'] > '2013-01-01') & (props_2013['created_on'] < '2014-01-01')]
props_2013_usd_month.index=props_2013_usd_month['created_on']
usd_month_2013 = props_2013_usd_month.groupby(pd.TimeGrouper(freq='M')).agg({'price_usd_per_m2': [np.mean]})
plt.plot(usd_month_2013)
```

```python
#Calculamos el valor promedio del m2 en dolares por mes para el año 2014
#Primero nos aseguramos que en el frame haya datos de solo el 2014

props_2014_usd_month = props_2014[(props_2014['created_on'] > '2014-01-01') & (props_2014['created_on'] < '2015-01-01')]
props_2014_usd_month.index=props_2014_usd_month['created_on']
usd_month_2014 = props_2014_usd_month.groupby(pd.TimeGrouper(freq='M')).agg({'price_usd_per_m2': [np.mean]})
plt.plot(usd_month_2014)
```

```python
#Calculamos el valor promedio del m2 en dolares por mes para el año 2015
#Primero nos aseguramos que en el frame haya datos de solo el 2015

props_2015_usd_month = props_2015[(props_2015['created_on'] > '2015-01-01') & (props_2015['created_on'] < '2016-01-01')]
props_2015_usd_month.index=props_2015_usd_month['created_on']
usd_month_2015 = props_2015_usd_month.groupby(pd.TimeGrouper(freq='M')).agg({'price_usd_per_m2': [np.mean]})
plt.plot(usd_month_2015)
```

```python
#Calculamos el valor promedio del m2 en dolares por mes para el año 2016
#Primero nos aseguramos que en el frame haya datos de solo el 2016

props_2016_usd_month = props_2016[(props_2016['created_on'] > '2016-01-01') & (props_2016['created_on'] < '2017-01-01')]
props_2016_usd_month.index=props_2016_usd_month['created_on']
usd_month_2016 = props_2016_usd_month.groupby(pd.TimeGrouper(freq='M')).agg({'price_usd_per_m2': [np.mean]})
plt.plot(usd_month_2016)
```

```python
#Calculamos el valor promedio del m2 en dolares por mes para el año 2017
#Primero nos aseguramos que en el frame haya datos de solo el 2017

props_2017_usd_month = props_2017[(props_2017['created_on'] > '2017-01-01') & (props_2017['created_on'] < '2018-01-01')]
props_2017_usd_month.index=props_2017_usd_month['created_on']
usd_month_2017 = props_2017_usd_month.groupby(pd.TimeGrouper(freq='M')).agg({'price_usd_per_m2': [np.mean]})
plt.plot(usd_month_2017)
```

```python
fig = plt.figure()
plt.plot(usd_month_2013)
plt.plot(usd_month_2014)
plt.plot(usd_month_2015)
plt.plot(usd_month_2016)
plt.plot(usd_month_2017)
fig.suptitle('Precio m2 en dolares por mes ', fontsize=20)
plt.ylabel('Precio USD', fontsize=14)
plt.xlabel('Mes', fontsize=16)
plt.legend()
plt.show()
#usd_month_2014.plot(kind='line',secondary_y=True)
#usd_month_2015.plot(kind='line',secondary_y=True)
#usd_month_2016.plot(kind='line',secondary_y=True)
#usd_month_2017.plot(kind='line',secondary_y=True)
```
